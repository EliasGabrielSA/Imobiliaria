## Etapa 3, valores medios de aluguel por tipo de imovel

```python
dados.head()
```

```python
medias_gerais = dados.groupby('Tipo').mean(numeric_only = True)
medias_gerais
```

```python
aluguel_medio = medias_gerais[['Valor']].sort_values('Valor')
aluguel_medio
```

## Etapa 4, removendo imóveis comerciais

```python
dados.Tipo.unique()
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala', 
                      'Prédio Inteiro', 'Loja/Salão', 
                      'Galpão/Depósito/Armazém', 
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
```

```python
imoveis_residenciais = dados.query('@imoveis_comerciais not in Tipo')
##imoveis_residenciais.plot(kind = 'barh', color = 'orange');
```

## Etapa 5, percentual de cada tipo de imovel na base de dados

```python
imoveis_residenciais.Tipo.unique()
```

```python
percentual_tipo = imoveis_residenciais.Tipo.value_counts(normalize = True).to_frame().sort_values('Tipo')
percentual_tipo.plot(kind = 'bar', color = 'orange', xlabel = 'Tipos', ylabel = 'Percentual')
```

## Etapa 6, Selecionando apenas imóveis do tipo apartamento

```python
apartamentos = imoveis_residenciais.query('Tipo == "Apartamento"')
apartamentos.head()
```
